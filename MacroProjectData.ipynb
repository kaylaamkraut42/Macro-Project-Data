```python
#SP500 (^GSPC)

import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Linear regression
#reindexing unemployment
un_rate = unrate.loc[444:920]
undf = pd.DataFrame(un_rate.get('UNRATE'))
spdf = pd.DataFrame(sp500.get('Adj Close'))
model = LinearRegression()
model.fit(spdf, undf)
y_pred = model.predict(spdf)

plt.scatter(spdf, undf, color='darkcyan')

plt.plot(spdf, y_pred, color='crimson')

plt.xlabel('S&P 500 Stock Prices (^GSPC)')
plt.ylabel('Unemployment Rate')
plt.title('S&P 500 Stock Prices and Unemployment Rate by Month')

plt.savefig('Macro_LinearRegression_1.png')
plt.show()


#Correlation coefficient calculation
undf_reset = undf.reset_index(drop=True)
spdf_reset = spdf.reset_index(drop=True)

joined_df = pd.concat([undf_reset, spdf_reset], axis=1)

corr1 = joined_df['^GSPC'].corr(joined_df['UNRATE'])
print("Correlation:", corr1)

spdf['pct_change'] = spdf['^GSPC'].pct_change()
undf['pct_change'] = undf['UNRATE'].pct_change()
```

```python
import matplotlib.dates as mdates

#Calculating percent change over year
spdf_year = spdf.iloc[11::12].reset_index(drop=True)
spdf_year['pct_change'] = spdf_year['^GSPC'].pct_change()

dates = pd.date_range(start="1985-01-01", end="2024-09-01", freq='MS')
datesY = pd.date_range(start="1984-12-01", end="2023-12-01", freq='YE')

#Graphing percent change of S&P on the left axis and unemployment on the right axis
y1 = spdf_year['pct_change']
y2 = undf['UNRATE']

fig, ax1 = plt.subplots(figsize=(20, 6))

bar_width = 350
ax1.bar(datesY, y1, width = bar_width,color = 'darkcyan', label='% Change S&P 500 Stock Prices')
ax1.set_xlabel('Year')
ax1.set_ylabel('S&P 500 Stock Prices % Change by Year', color='darkcyan')
ax1.tick_params(axis='y', labelcolor='darkcyan')

ax2 = ax1.twinx()
ax2.plot(dates, y2, 'r', color = 'crimson', label='Unemployment Rate')
ax2.set_ylabel('Unemployment Rate', color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
ax1.xaxis.set_major_locator(mdates.YearLocator(1))
plt.xticks(rotation=45)

fig.suptitle('S&P 500 Stock Prices (^GSPC)')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

#Show the plot
plt.tight_layout()
plt.show()
```

```python
#Percent change of S&P by Month
dates = pd.date_range(start="1985-01-01", end="2024-09-01", freq='MS')

y1 = spdf['pct_change']
y2 = undf['UNRATE']

fig, ax1 = plt.subplots(figsize=(20, 6))

bar_width = 20 
ax1.bar(dates, y1, width = bar_width, color='darkcyan', label='% Change in S&P 500 Stock Prices')
ax1.set_xlabel('Month')
ax1.set_ylabel('S&P 500 Stock Prices % Change by Month', color='darkcyan')
ax1.tick_params(axis='y', labelcolor='darkcyan')

ax2 = ax1.twinx()
ax2.plot(dates, y2, 'r', color = 'crimson', label='% Change in Unemployment Rate')
ax2.set_ylabel('Unemployment Rate % Change', color='crimson')
ax2.tick_params(axis='y', labelcolor='crimson')

ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
ax1.xaxis.set_major_locator(mdates.YearLocator(2))
plt.xticks(rotation=45)

fig.suptitle('S&P Stock Prices (^GSPC) by Month')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

# Data for the grid
cell_data = [
    {"color": "plum", "text": "Communication\nServices: -0.43"},
    {"color": "orange", "text": "Consumer\nDiscretionary: -0.44"},
    {"color": "blueviolet", "text": "Consumer\nStaples: -0.45"},
    {"color": "mediumseagreen", "text": "Energy: -0.37"},
    {"color": "lightskyblue", "text": "Financial: -0.63"},
    {"color": "lightsalmon", "text": "Health Care: -0.46"},
    {"color": "royalblue", "text": "Industrials: -0.52"},
    {"color": "hotpink", "text": "Technology: -0.41"},
    {"color": "coral", "text": "Materials: -0.46"},
    {"color": "lawngreen", "text": "Real Estate: -0.53"},
    {"color": "gold", "text": "Utilities: -0.48"},
]

# Create figure
fig, ax = plt.subplots(figsize=(30, 10))

# Grid dimensions
rows, cols = 3, 4
cell_width, cell_height = 1, 1

# Font settings
font_properties = {
    'family': 'monospace',  # Font family (e.g., 'serif', 'sans-serif')
    'weight': 'normal',   # Font weight (e.g., 'normal', 'bold')
    'size': 14          # Font size
}

# Draw grid
for i, cell in enumerate(cell_data):
    row = i // cols
    col = i % cols

    # Draw rectangle for each cell
    rect = Rectangle((col, rows - row - 1), cell_width, cell_height, facecolor=cell["color"], edgecolor="black")
    ax.add_patch(rect)

    # Add text to each cell with custom font
    ax.text(
        col + 0.5, rows - row - 0.5, cell["text"], 
        color="black", ha="center", va="center", fontsize=font_properties['size'],
        fontweight=font_properties['weight'], fontfamily=font_properties['family']
    )

# Set limits and aspect
ax.set_xlim(0, cols)
ax.set_ylim(0, rows)
ax.set_aspect("equal")
ax.axis("off")  # Hide axes
plt.savefig('Macro_CorrCoeffGrid_2.png')

# Show the plot
plt.show()
```
